```python
df_limpio = df.dropna()
```

```python
print(f"Filas de datos original: {conteo_df_original}")
print(f"Filas de datos limpios: {conteo_df_limpio}")
```

```python
df.isnull().any().any()
```

```python
df_limpio.isnull().any().any()
```

```python
tipos = df.columns.to_series().groupby(df.dtypes).groups
ctext = tipos[np.dtype('object')]
len(ctext)
```

```python
tipos = df_limpio.columns.to_series().groupby(df_limpio.dtypes).groups
ctext = tipos[np.dtype('object')]
len(ctext)
```

```python
columnas = df.columns
cnum = list(set(columnas) - set(ctext))
print(len(cnum))
```

```python
columnas = df_limpio.columns
cnum = list(set(columnas) - set(ctext))
len(cnum)
```

```python
# Completando valores faltantes datos cuantititavos
for columna in cnum:
  mean = df[columna].mean()
  df[columna] = df[columna].fillna(mean)
```

```python
for columna in ctext:
  mode = df[columna].mode()[0]
  df[columna] = df[columna].fillna(mode)
```

```python
df.isnull().any().any()
```

```python
donantes = df.groupby('DONOR_FLAG').IDX.count()

labels = [ 'Donante\n' + str(round(x * 1.0 / donantes.sum() *
100.0, 2)) + '%' for x in donantes ]
labels[0] = 'No ' + labels[0]
plt.pie(donantes, labels=labels)
plt.title('Porcion de donantes')
plt.show()
```

```python
data_donantes = df[df.DONOR_AMOUNT > 0]
data_donantes.shape
```

```python
monto_segm = pd.cut(data_donantes['DONOR_AMOUNT'], [0, 10, 20, 30, 40, 50, 60, 100, 200])

plot = pd.value_counts(monto_segm).plot(kind='bar', title='Montos de donación')
plot.set_ylabel('Nro. de donantes')
plot.set_xlabel('Rango de montos')
plt.show()
```

```python
pd.value_counts(monto_segm)
```

```python
data_donantes['DONOR_AMOUNT'].mean()
```

```python
plt.boxplot(list(data_donantes['DONOR_AMOUNT']))
plt.title('Montos de donación')
plt.show()
```

```python
data_donantes.groupby('GENDER').size().plot(kind='bar')
plt.title('Distribución por género')
plt.show()
```

```python
data_donantes[(data_donantes.DONOR_AMOUNT <= 50)
& (data_donantes.GENDER.isin(['F', 'M']))][['DONOR_AMOUNT', 'GENDER']].boxplot(by='GENDER')
plt.title('Donantes según género')
plt.show()
```
