```python
#Aca extraemos los 10 directores mas repetidos y los que no encajan en este grupo los categorizamos como otros, esto sirve, ya que al ser 548 director
#esto se engrandece mucho y aplicamos una simplificacion de datos
Top10_Directores = df['Director'].value_counts().nlargest(10).index
df['director_agrupado'] = df['Director'].apply(lambda x: x if x in Top10_Directores else 'Otros')
df['director_agrupado'].value_counts()
```
